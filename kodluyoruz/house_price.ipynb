```python
variable_balances = {}
for column in df.columns:
    frequencies = df[column].value_counts(normalize=True) * 100
    variable_balances[column] = frequencies

#Sonuçları yazdır
for variable, frequencies in variable_balances.items():
    print(f"Denge Dağılımı - {variable}")
    print(frequencies)
    print()
```
